```python
import matplotlib.pylab as plt
import numpy as np
from IPython.display import clear_output
```

```python
%%bash
g++ pendulum.cpp
./a.out 10000000 > data.txt
```

```python
da = np.genfromtxt("data.txt")
t = da[:,0]
theta = da[:,1]
omega = da[:,2]
plt.plot(t[:10000], theta[:10000])
plt.xlabel("tiempo")
plt.ylabel("$\theta$")
plt.savefig("image.png")
plt.show()
```

```python
for i in range(10, len(theta), 100):
    plt.plot(theta[:i], omega[:i], ".")
    plt.ylim(-3,1.5)
    plt.xlim(-3.5,3.5)
    plt.xlabel(theta)
    plt.ylabel(omega)
    plt.savefig(str(i).zfill(4)+'.png')
    plt.show()
    if i>9910:
        break;
    
```
